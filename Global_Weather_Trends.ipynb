## Global Weather Data Exploration

```python
from pylab import rcParams
import statsmodels.api as sm

rcParams['figure.figsize'] = 11, 9
decomposed_avg_temp = sm.tsa.seasonal_decompose(global_data['avg_temp'],freq=1)
figure = decomposed_avg_temp.plot()
plt.show()
```

You can clearly see an upward trend in global avg temperatures.

```python
rcParams['figure.figsize'] = 16, 6
white_noise = np.random.normal(loc=0, scale=1, size=1000)
plt.plot(white_noise)
```

```python
# The original non-stationary plot

decomposed_avg_temp.trend.plot()
```

```python
# New stationary plot

decomposed_avg_temp.trend.diff().plot()
```

## NYC Moving Average Analysis

```python
nyc = city_data[city_data['city'] == 'New York']
print(nyc)
```

```python
nyc.dropna()

nyc = nyc.drop(nyc[nyc.avg_temp < 0.5].index)


nyc['moving_avg_nyc'] = nyc['avg_temp'].rolling(window=5).mean()
```

```python
nyc2 = nyc.merge(global_data, left_on = 'year', right_on = 'year', how = 'left')
```

```python
nyc2['moving_avg_global'] = nyc2['avg_temp_y'].rolling(window=5).mean()
```

```python
nyc2.dropna()
```

```python
nyc2['moving_avg_var'] = (nyc2['moving_avg_nyc']- nyc2['moving_avg_global'])/nyc2['moving_avg_global']
```

```python
x = nyc2['year']

plt.plot(x, nyc2['moving_avg_nyc'], 'b--', label = 'nyc_5yr_rolling_avg', linewidth = 3.0)
plt.plot(x, nyc2['moving_avg_global'], 'c--', label = 'global_5yr_rolling_avg', linewidth = 3.0)
plt.fill_between(x, nyc2['moving_avg_global'], nyc2['moving_avg_nyc'], facecolor = 'green')

plt.xlabel('Year')
plt.ylabel('Temp')
plt.title('NYC vs Global 5 Yr Rolling Average Temperature')
plt.legend()
```

```python
# Calculate Correlation Coefficient between average NYC and Global temperatures 

nyc2['moving_avg_nyc'].corr(nyc2['moving_avg_global'], method = 'pearson')
```

```python
nyc2['avg_temp_x'].corr(nyc2['avg_temp_y'], method = 'pearson')
```

Percent variance between the two:

```python
plt.plot(x, nyc2['moving_avg_nyc'], 'b--', label = 'percent_variation', linewidth = 3.0)

plt.xlabel('Year')
plt.ylabel('% Variance')
plt.title('Percent Variance NYC vs Global Average Temperature')
plt.legend()
```

## Initial Takeways

- NYC and Global 5-year rolling average temperatures have a positive correlation coefficient of 81%.
